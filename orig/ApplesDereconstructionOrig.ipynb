```python
def thing(U, s, Vh, n):
    accum = np.zeros_like(U)
    for i in range(0, n):
        accum += np.outer(U[:,i], Vh[i,:]) * s[i]
    return accum
```

```python
plt.imshow(thing(U,s,Vh,10), cmap='gray')
```

```python
thing(U,s,Vh,255) - thing(U,s,Vh,254)
```
